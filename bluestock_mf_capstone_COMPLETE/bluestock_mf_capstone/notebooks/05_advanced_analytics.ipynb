```
import pandas as pd
from pathlib import Path

ROOT    = Path().resolve().parent
REPORTS = ROOT / 'reports'

# all output files from advanced analytics
output_files = {
    'var_cvar_report.csv'  : 'VaR and CVaR for all 40 funds',
    'fund_scorecard.csv'   : 'Composite scorecard (0-100) for all 40 funds',
    'alpha_beta.csv'       : 'Alpha and Beta vs Nifty 100 for all funds',
    'cohort_analysis.csv'  : 'Investor cohort analysis by first transaction year',
    'sip_continuity.csv'   : 'SIP continuity analysis with at-risk flags',
    'sector_hhi.csv'       : 'Herfindahl-Hirschman Index per fund',
}

print("Advanced Analytics Output Files:")
for fname, desc in output_files.items():
    fpath = REPORTS / fname
    if fpath.exists():
        df = pd.read_csv(fpath)
        print(f"  {fname}: {len(df)} rows - {desc}")
    else:
        print(f"  {fname}: NOT FOUND - run 06_Advanced_Analytics.ipynb first")
```

```
# quick summary of key results
try:
    sc  = pd.read_csv(REPORTS / 'fund_scorecard.csv')
    var = pd.read_csv(REPORTS / 'var_cvar_report.csv')
    ab  = pd.read_csv(REPORTS / 'alpha_beta.csv')

    print("TOP 5 FUNDS - Composite Scorecard:")
    print(sc.head(5)[['scheme_name','composite_score']].to_string(index=False))
    print()
    print("HIGHEST RISK (VaR 95%):")
    print(var.head(3)[['scheme_name','var_95_pct','cvar_95_pct']].to_string(index=False))
    print()
    print("HIGHEST ALPHA (vs Nifty 100):")
    print(ab.head(3)[['scheme_name','alpha_ann','beta']].to_string(index=False))
except FileNotFoundError:
    print("Run 06_Advanced_Analytics.ipynb first to generate outputs.")
```
