## 0. Setup – Imports & Data Loading

```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns
import plotly.graph_objects as go
import plotly.express as px
import warnings

warnings.filterwarnings("ignore")
matplotlib.rcParams["figure.dpi"] = 120
sns.set_theme(style="whitegrid")

PROCESSED = "../data/processed"
CHARTS    = "charts"

# Load all cleaned datasets
nav   = pd.read_csv(f"{PROCESSED}/02_nav_history_clean.csv",   parse_dates=["date"])
aum   = pd.read_csv(f"{PROCESSED}/03_aum_by_fund_house_clean.csv", parse_dates=["date"])
sip   = pd.read_csv(f"{PROCESSED}/04_monthly_sip_inflows_clean.csv", parse_dates=["month"])
cat   = pd.read_csv(f"{PROCESSED}/05_category_inflows_clean.csv",   parse_dates=["month"])
folio = pd.read_csv(f"{PROCESSED}/06_industry_folio_count_clean.csv", parse_dates=["month"])
perf  = pd.read_csv(f"{PROCESSED}/07_scheme_performance_clean.csv")
txn   = pd.read_csv(f"{PROCESSED}/08_investor_transactions_clean.csv", parse_dates=["transaction_date"])
ph    = pd.read_csv(f"{PROCESSED}/09_portfolio_holdings_clean.csv")
fm    = pd.read_csv(f"{PROCESSED}/01_fund_master_clean.csv")

print("All datasets loaded successfully")
print(f"  NAV rows        : {len(nav):,}")
print(f"  Transactions    : {len(txn):,}")
print(f"  SIP months      : {len(sip)}")
print(f"  Funds           : {len(fm)}")
```

## 1. NAV Trend Analysis – All 40 Schemes (2022–2026)

> **Finding 1:** Equity fund NAVs grew ~2.3× between Jan 2022 and Dec 2025, with a strong bull run through 2023 followed by a short correction in mid-2024 before resuming the uptrend — confirming the resilience of Indian equity markets over the 4-year period. *(Chart 1)*

```python
from IPython.display import Image
Image(f"{CHARTS}/chart01_nav_trend.png")
```

## 2. AUM Growth by Fund House (2022–2025)

> **Finding 2:** SBI Mutual Fund maintains clear AUM dominance at ₹12.5 Lakh Crore, nearly 3× its closest competitor ICICI Prudential, highlighting the significant market concentration among top-3 AMCs. *(Chart 2)*

```python
Image(f"{CHARTS}/chart02_aum_bar.png")
```

## 3. SIP Inflow Time-Series (Jan 2022 – Dec 2025)

> **Finding 3:** Monthly SIP inflows grew from ₹11,517 Cr in Jan 2022 to an all-time high of ₹31,002 Cr in Dec 2025, a 169% increase in 4 years, reflecting deepening retail investor participation in mutual funds. *(Chart 3)*

```python
Image(f"{CHARTS}/chart03_sip_timeseries.png")
```

## 4. Category Inflow Heatmap

> **Finding 4:** Equity funds (Large Cap, Mid Cap, Small Cap) consistently attract the highest net inflows year-round, while Debt fund inflows show strong seasonality — spiking at financial year-end (March) due to institutional demand. *(Chart 4)*

```python
Image(f"{CHARTS}/chart04_category_heatmap.png")
```

## 5. Investor Demographics

> **Finding 5:** The 25–34 and 35–44 age groups together account for over 60% of all transactions, and male investors make up ~65% of the investor base — indicating that younger urban males are the primary drivers of mutual fund growth. *(Chart 5)*

```python
Image(f"{CHARTS}/chart05_demographics.png")
```

## 6. Geographic Distribution

> **Finding 6:** Punjab and Madhya Pradesh lead in transaction count, but Maharashtra and Gujarat dominate by transaction value — suggesting that high-net-worth investors in Tier-1 cities drive the majority of AUM despite lower transaction frequency. *(Chart 6)*

```python
Image(f"{CHARTS}/chart06_geographic.png")
```

## 7. Folio Count Growth

> **Finding 7:** Industry folios nearly doubled from 13.26 Crore in Jan 2022 to 26.12 Crore by Dec 2025, crossing the 20-Crore milestone in 2024 — demonstrating the rapid expansion of the mutual fund investor base in India. *(Chart 7)*

```python
Image(f"{CHARTS}/chart07_folio_growth.png")
```

## 8. NAV Return Correlation Matrix

> **Finding 8:** Funds within the same category (e.g., Large Cap vs Large Cap) show near-perfect return correlation (0.95+), while Debt and Equity funds show near-zero or negative correlation, confirming their value as portfolio diversifiers. *(Chart 8)*

## 9. Sector Allocation – Equity Funds
